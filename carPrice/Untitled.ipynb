```python
#aim preprocessing the data to
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
data=pd.read_csv(filepath,names=headers)
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.replace("?",np.nan,inplace=True)
```

```python
data
```

```python
missing_data=data.isnull()
```

```python
missing_data
```

```python
for columns in missing_data.columns.values.tolist():
    print(missing_data[columns].value_counts(sort=True,ascending=True))
```

```python
avg_loss=data["normalized-losses"].astype("float").mean(axis=0)
data["normalized-losses"].replace(np.nan,avg_loss,inplace=True)
```

```python
mean=data[["bore","stroke","horsepower","peak-rpm"]].astype("float").mean(axis=0)
```

```python
data["bore"].replace(np.nan,mean[0],inplace=True)
data["stroke"].replace(np.nan,mean[1],inplace=True)
data["horsepower"].replace(np.nan,mean[2],inplace=True)
data["peak-rpm"].replace(np.nan,mean[3],inplace=True)
```

```python
data["num-of-doors"].replace(np.nan,4,inplace=True)
```

```python
data.dropna(subset=['price'],axis=0,inplace=True)
data.reset_index(drop=True,inplace=True)
```

```python
data.dtypes
```

```python
data[["price","bore","stroke","peak-rpm"]]=data[["price","bore","stroke","peak-rpm"]].astype("float",inplace=True)
data[["normalized-losses","horsepower"]]=data[["normalized-losses","horsepower"]].astype(int)
```

```python
bins=np.linspace(min(data["horsepower"]),max(data["horsepower"]),4)
group_name=["low","medium","high"]
data["horsepower binned"]=pd.cut(data["horsepower"],bins,labels=group_name,include_lowest=True)

```

```python
data["horsepower binned"].value_counts()
```

```python
dummy=pd.get_dummies(data["fuel-type"])
print(dummy.head())
```

```python
data=pd.concat([data,dummy],axis=1)
data.drop(["fuel-type","horsepower"],axis=1,inplace=True)
```

```python
dummy2=pd.get_dummies(data["aspiration"])
data=pd.concat([data,dummy2],axis=1)
data.drop("aspiration",axis=1,inplace=True)
```

```python
value_data=data[["wheel-base","length","width","curb-weight","engine-size","bore","city-mpg",'highway-mpg']]
```

```python
value_data.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_reg=LinearRegression()
```

```python
linear_reg.fit(value_data,data['price'])
```

```python
plt.figure(figsize=(12,10))
sns.distplot(Yhat,label="fitted value")
sns.distplot(data["price"],label="actual value")
plt.legend(loc='best')
plt.show()
plt.close()
```
